```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Coverage = pd.read_csv("ID_Coverage.csv")
Comparison = pd.read_csv("Comparison.csv")
Coverage = Coverage[Coverage["Track Length (CME)"]==99]
cov = Coverage.to_numpy()
```

```python
plt.bar(cov[:,0].astype(int).astype(str), 100*(cov[:,2]/cov[:,1]), width = 0.5)
plt.xlabel("ID (CME)")
plt.ylabel("Coverage (%)")
plt.title("Coverage for Track Length = 99")
plt.show()
plt.savefig("Coverage TL 99.png")
```

```python
plt.hist(Comparison["Distance"], bins = round(max(Comparison["Distance"])), density=True)
plt.xlim(0,50)
plt.xticks(np.arange(0, 50, 1.0))
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.show()
plt.savefig("Histogram Distance.png")
```

```python
plt.scatter(Comparison["Distance"], Comparison["FI (CME)"], s = 0.5)
plt.xlabel("Distance")
plt.ylabel("FI (CME)")
plt.show()
plt.savefig("FI CME vs Distance.png")
```

```python
plt.scatter(Comparison["FI (Dino)"], Comparison["FI (CME)"], s = 0.01)
plt.xlabel("FI (Dino)")
plt.ylabel("FI (CME)")
plt.show()
plt.savefig("FI CME vs CI Dino.png")
```
